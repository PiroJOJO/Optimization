```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return 0.1 * (x**3 - x) + 1
```

```python
X = np.linspace(-1, 1, 1000, endpoint = False)
Y = np.array([func(t) for t in X])
```

```python

plt.xlabel('Время, с')
plt.ylabel('Амплитуда, м')
plt.title('f(x)')
plt.plot(X, Y)
plt.show()

```

```python
def func_(x):
    return 0.1 * (3*x**2 - 1) 
```

```python
Y_ = np.array([func_(t) for t in X])
L = np.max(Y_)
print(L)
```

```python
a = 0 
b = 1000 - 1
print(X[-1])
```

```python
x0 = 1 / (2 * L)* (Y[a] - Y[b] + L * (X[a] + X[b]))
y0 = 1 / 2 * (Y[a] + Y[b] + L * (X[a] - X[b]))
print(x0)
print(y0)
p0=[]
for x in X:
    if x<=x0:
        p0.append(Y[a] - L * (x - X[a]))
    else:
        p0.append(Y[b] + L * (x - X[b]))
```

```python
plt.xlabel('Время, с')
plt.ylabel('Амплитуда, м')
plt.title('f(x)')
plt.plot(X,Y)
plt.plot(X,p0,'r--')
plt.show()
```

```python

e = 0.015
x_min = x0
y_min = y0
points_min = [[x0],[y0]]
f_max = max(Y)
delta = 1 / (2 *L)*(func(x_min) - y0)
x_ = x_min - delta
x__ = x_min + delta
y_ = 1 / 2 *(func(x_min) + y0)

points_min[0][0] = x__
points_min[1][0] = y_
points_min[0].append(x_)
points_min[1].append(y_)

error = func(x0) - y0
while(error > e):
    y_min = min(points_min[1])
    id_min = points_min[1].index(y_min)
    x_min = points_min[0][id_min]

    delta = 1 / (2 *L)*(func(x_min) - y_min)
    x_ = x_min - delta
    x__ = x_min + delta
    y_ = float(1 / 2 *(func(x_min) + y_min))

    points_min[0].append(x_)
    points_min[1].append(y_)

    points_min[0].append(x__)
    points_min[1].append(y_)

    points_min[1][id_min] = f_max
    error = func(x_min) - y_min
    plt.plot()

print('Predict x: ', x_min)
print('Predict y: ',y_min)
print('True_answer', min(Y))
```
